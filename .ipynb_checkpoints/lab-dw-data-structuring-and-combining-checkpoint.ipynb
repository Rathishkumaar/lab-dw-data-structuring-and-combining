# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df = pd.read_csv(url)

# Display the first few rows of the dataset
print(df.head())

# Display the basic information of the dataset
print(df.info())

# Display the summary statistics for numerical columns
print(df.describe().transpose())

# Display the summary statistics for categorical columns
print(df.describe(include=['object']).transpose())

# Check for missing values
print(df.isnull().sum())

# Option 1: Drop rows with missing values
df_cleaned = df.dropna()

# Option 2: Fill missing values with appropriate values (mean, median, mode, etc.)
# For demonstration, let's fill numerical columns with their mean and categorical columns with their mode
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

for col in numerical_columns:
    df[col].fillna(df[col].mean(), inplace=True)

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print(df.isnull().sum())  # Check if missing values are handled
```

```python
# Convert columns to appropriate data types if needed
# For example, if 'Income' was read as an object, we should convert it to a numerical type
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')

# Ensure date columns are in datetime format if applicable
# df['SomeDateColumn'] = pd.to_datetime(df['SomeDateColumn'], errors='coerce')

print(df.dtypes)
```

```python
# Rename columns for consistency and readability
df.rename(columns={
    'Customer': 'CustomerID',
    'ST': 'State',
    'Gender': 'Gender',
    'Education': 'Education',
    'Customer Lifetime Value': 'CustomerLifetimeValue',
    'Income': 'Income',
    'Monthly Premium Auto': 'MonthlyPremiumAuto',
    'Number of Open Complaints': 'OpenComplaints',
    'Policy Type': 'PolicyType',
    'Vehicle Class': 'VehicleClass',
    'Total Claim Amount': 'TotalClaimAmount',
    # Add any additional columns that need renaming
}, inplace=True)

print(df.columns)
```

```python
# Example: Create a new column for 'Annual Premium Auto'
df['AnnualPremiumAuto'] = df['MonthlyPremiumAuto'] * 12

print(df.head())
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_marketing_customer_analysis.csv', index=False)
```
